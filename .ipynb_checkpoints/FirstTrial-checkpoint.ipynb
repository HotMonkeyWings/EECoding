```python
# Statewise Evapotranspiration Heatmap
df = pd.read_csv("02_Terra Net Evapotranspiration 8-Day Global 500m_2011_Dist.csv")

df['date'] = pd.to_datetime(df['date'],format='%d-%m-%Y').dt.date
# df = df[df['date'] != pd.to_datetime('01-01-2001',format='%d-%m-%Y').dt.date]
heatmap1_data = pd.pivot_table(df, values='mean', index=['ST_NM'], columns='date')
plt.figure(figsize=(12, 12))

sns.heatmap(heatmap1_data, cmap="YlGnBu")


#Hard to observe due to seasons.
```

```python
#For Q1
heatmap1_data = pd.pivot_table(df, values='mean', index=['ST_NM'], columns='date')
df2 = heatmap1_data.copy()
for x in df2.columns:
    if(x.month != 1):
        df2 = df2.drop(columns=[x])
# print(df2)
# plt.figure(figsize=(12, 12))
# print(df2.columns)
# for ind in df2:
#     print(df2[df2.columns[0]][ind])
df2.index

for ind in df2.index:
    x = []
    y = []
    for col in df2.columns:
        #print(col,ind,df2[col][ind])
        x.append(col)
        y.append(df2[col][ind])
    x_temp = []
    for val in range(len(x)):
        x_temp.append(val)
    x_temp = np.array(x_temp)
    
    y = np.array(y)
    m, b = np.polyfit(x_temp,y,1)
    plt.plot(x,y,'o')
    plt.plot(x,m*x_temp+b)
    plt.title(ind)
    plt.show()
```
